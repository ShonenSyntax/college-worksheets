## Getting list of datasets available in pycaret

```python
from pycaret.datasets import get_data
dataSets = get_data('index')

# PyCaret has a module called datasets which contains ready-to-use datasets.
# get_data is a function that fetches a dataset from PyCaret’s repository.

#'index' is not a Python keyword, it’s just a string that the PyCaret developers
# chose to trigger a special behavior. (It has been defined in the Source Code.)
```

## Loading the Diabetes Dataset

```python
diabetesDataSet = get_data('diabetes')
```

## Explorin the Dataset

```python
diabetesDataSet.describe()
#Get the statistical summary of the dataset
```

## Get the dimensions of the Dataset

```python
print("diabetesDataSet.shape -->", diabetesDataSet.shape)

# Print the number of rows (axis 0)
print("Rows  -->", diabetesDataSet.shape[0])
# Print the number of columns (axis 1)
print("Columns -->", diabetesDataSet.shape[1])
```

## Visualising the Diabetes Dataset

### Histogram Plot

```python
import matplotlib.pyplot as plt

x = diabetesDataSet['Age (years)']

plt.xlabel('Age')
plt.ylabel('Frequency')

plt.hist(x,color = 'green', edgecolor = 'black', bins = 50)
```

### ScatterPlot

```python
x = diabetesDataSet['Number of times pregnant']
y = diabetesDataSet['Body mass index (weight in kg/(height in m)^2)']

plt.xlabel('Prego Frequency')
plt.ylabel('BMI')
plt.scatter(x, y, marker= '+', color = 'purple')

plt.show()
```
