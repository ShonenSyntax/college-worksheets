# Task

### Generate synthetic monthly temperature data for two cities using NumPy, 
### store it in a Pandas DataFrame, and use Matplotlib to create a line plot,
### bar chart, and scatter plot.

```python
import numpy as np

#Generating Synthetic monthly temperature data
xdata = np.random.normal(24,3,12)
ydata = np.random.normal(34,3,12)
xtemp = np.round(xdata,1)
ytemp = np.round(ydata,1)
print(xtemp)
print(ytemp)
```

```python
import pandas as pd

#Storing the data in a Pandas DataFrame

months = ['Jan','Feb','Mar','Apr','May','Jun','Jul',
          'Aug','Sept','Oct','Nov','Dec']
data = {
    'Months': months,
    'City_A': xtemp,
    'City_B': ytemp
}

df = pd.DataFrame(data)
print(df)
```

```python
import matplotlib.pyplot as plt

#Creating the LinePlot Illustration
# plt.plot(x_values, y_values, options...)

plt.plot(months, xtemp, marker = 'o', linestyle = 'dotted', label = 'City_A')
plt.plot(months, ytemp, marker = '*', linestyle = 'dashed', label = 'City_B')

plt.title('Temperature Data - LinePlot')
plt.xlabel('Months')
plt.ylabel("Temperature (°C)")
plt.tight_layout()
plt.legend()
plt.show()
```

```python
#Creating the BarChart Illustration
# plt.bar(categories_or_positions, heights, options...)

x = np.arange(len(months))
width = 0.35

plt.bar(x-width/2,xtemp,width = 0.35, color = 'blue', label = 'City_A')
plt.bar(x+width/2,ytemp, width = 0.35, color = 'orange', label = 'City_B')

plt.xticks(x,months)
plt.title('Temperature Data - BarChart')
plt.xlabel('Months')
plt.ylabel("Temperature (°C)")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
#Creating the ScatterPlot Illustration
#plt.scatter(x_values, y_values, options...)

plt.scatter(months, xtemp, label= 'City_A', marker = 'o')
plt.scatter(months, ytemp, label = 'City_B', marker = '*')
plt.title("Temperature Data - Scatter Plot")
plt.xlabel('Months')
plt.ylabel("Temperature (°C)")
plt.legend()
plt.tight_layout()
plt.show()
```
