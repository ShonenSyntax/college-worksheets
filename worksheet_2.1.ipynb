```python
#Importing Pycaret and checking it's version
import pycaret
print(pycaret.__version__)
```

```python
#loading the dataSets
import pandas as pd
from pycaret.datasets import get_data
dataSets = get_data('index')
```

```python
titanicDataSet = get_data('titanic')
```

```python
## Print average age and fare and no. of passengers
print('Total no. of Passengers-->',titanicDataSet.shape[0])
print("Mean[Age]-->",titanicDataSet['Age'].mean())
print("Mean[Fare]-->",titanicDataSet['Fare'].mean())
```

```python
#Checking for missing Values
titanicDataSet.isnull().sum()
```

## Visualisations

### Histogram

```python
import matplotlib.pyplot as plt

x = titanicDataSet['Age']
plt.xlabel('Ages')
plt.ylabel('Frequency')
plt.title('Age Distribution')
plt.hist(x, color = 'green', edgecolor = 'black')
plt.tight_layout()
plt.show()
```

### PieChart

```python
count = titanicDataSet[titanicDataSet['Survived'] == 1]['Sex'].value_counts()
count.plot(kind= 'pie',colors= ['pink','lightblue'], startangle = 90)
plt.legend([f"female: {count['female']}",f"male: {count['male']}"])
plt.title('Titanic Survival(males vs females)')
plt.ylabel("")   #remove default y-label
plt.show()
```

### BarChart

```python
pd.crosstab(titanicDataSet['Pclass'],titanicDataSet['Survived']).plot(kind= 'bar', color = ['red','green'])

plt.title("Survival by Passenger Class on the Titanic")
plt.xlabel("Passenger Class")
plt.ylabel("Number of Passengers")
plt.legend(["Did Not Survive", "Survived"])
plt.show()
```
