## *Data Set Injestion*

```python
import pandas as pd

data = pd.read_csv('data.csv')
df = pd.DataFrame(data)
print(df.to_string())
```

### *Display Basic Information*

```python
print(df.info())
```

### *Summary Statistics*

```python
print(df.describe())
```

### *Check for Missing Values*

```python
print(df.isnull().sum())
```

### *Correlation Matrix*

```python
df.corr()

```

### *Heatmap of Correlations*

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))  #Create a new empty canvas that is 8 inches wide and 6 inches tall. Any plots I draw next will go onto this canvas
sns.heatmap(df.corr(), annot= True, cmap= 'coolwarm')
plt.title('Correlation Heatmap')
plt.show() 

#Because Saborns plotting functions (sns.heatmap, sns.scatterplot, etc.) 
#draw directly onto the current Matplotlib figure.

#So the sequence is:
#plt.figure(figsize=(8, 6))  Prepare canvas (size 8x6).
#sns.heatmap(...)  Draw the heatmap on that prepared canvas.
```

### *Histograms for Numeric Columns*

```python
df.hist(figsize=(6,6), color= 'green', edgecolor= 'black') #This tells pandas: For every numeric column in my DataFrame, draw a histogram.
plt.suptitle('Histograms for the Numeric Columns')  #adds a title for the entire figure (suptitle = "super title")
plt.tight_layout()
plt.show()
```

### *BoxPlot for Outlier Detection*

```python
sns.boxplot(data=df)
plt.title('Box Plot for all the Numeric Columns')
plt.show()
```

### *Pairplot for Feature Relationships*

```python
sns.pairplot(df)
plt.suptitle('Pairplots of the Dataset', y = 1.02)
plt.show()
```

### *ScatterPlot (Duration vs Calories)*

```python
plt.scatter(df['Duration'], df['Calories'], color= 'green')
plt.title('Duration vs Calories')
plt.xlabel('Duration')
plt.ylabel('Calories')
plt.grid()
plt.show()
```
